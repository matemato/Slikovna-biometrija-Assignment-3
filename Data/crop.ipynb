```python
%matplotlib inline

import pandas as pd
import cv2
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("ids.csv", header=None, names=["Name", "Class"])
```

```python
df.head()
```

```python
test_df = df.iloc[0:250,:]
train_df = df.iloc[250:,:].reset_index(drop=True)
```

```python
TRAIN_DIR = './Train'
TEST_DIR = './Test'

train_img = os.path.join(TRAIN_DIR, 'Images')
train_labels = os.path.join(TRAIN_DIR, 'labels')
train_segmentation = os.path.join(TRAIN_DIR, 'Segmentation')
train_classes = os.path.join(TRAIN_DIR, 'Classes')

test_img = os.path.join(TEST_DIR, 'Images')
test_labels = os.path.join(TEST_DIR, 'labels')
test_segmentation = os.path.join(TEST_DIR, 'Segmentation')
test_classes = os.path.join(TEST_DIR, 'Classes')
```

```python
for i, img in enumerate(sorted(os.listdir(test_img))):
    print(img)
    print(f'{test_img}/{img}')
    image = cv2.imread(f'{test_img}\{img}')
    segmentation = cv2.imread(f'{test_segmentation}\{img}')
    segmentation = segmentation[12:372,:]
    segmentation_vis = segmentation * 255
    img_class = test_df['Class'][i]
    try:
        file = open(f'{test_labels}/{str(i+1).zfill(4)}.txt', 'r')
        for line in file.readlines():
            l_arr = line.split(" ")[1:5]
            l_arr[2] = int(480 * float(l_arr[2]))
            l_arr[3] = int(360 * float(l_arr[3]))
            l_arr[0] = int(float(l_arr[0])*480 - int(l_arr[2]/2))
            l_arr[1] = int(float(l_arr[1])*360 - int(l_arr[3]/2))
            x = l_arr[0]
            y = l_arr[1]
            w = l_arr[2]
            h = l_arr[3]
            mask = np.zeros_like(image)
            # print(segmentation.shape)
            # print(image.shape)
            ear = image * segmentation
            ear = ear[y:y+h,x:x+w]
            # plt.imshow(cv2.cvtColor(ear, cv2.COLOR_BGR2RGB))
            # plt.show()
            try:
                os.mkdir(test_classes + "/" + str(img_class))
            except:
                pass
            if np.sum(ear) == 0:
                print("NO EAR")
            else:
                cv2.imwrite(f'{test_classes}/{str(img_class)}/{len(os.listdir(test_classes + "/" + str(img_class)))}.png', ear)
            
        # plt.imshow(segmentation_vis)
        # plt.show()
        print(f'Class: {img_class}, image: {i+1}')
        file.close()
    except Exception:
        print(Exception)
    
    # if i == 2: break
```
